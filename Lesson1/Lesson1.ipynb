Работа с данными в Pandas
Задание 1


```python
import pandas as pd
authors=pd.DataFrame({'author_id':[1,2,3], 'author_name':['Тургенев', 'Чехов', 'Островский']}, columns=['author_id','author_name'])
authors
```

```python
books=pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3],'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],'price':[450, 300, 350, 500, 450, 370, 290]},columns=['author_id','book_title','price'])
books
```

Задание 2


```python
authors_price=pd.merge(authors, books, on='author_id', how='inner')
authors_price
```

Задание 3

```python
top5=authors_price.nlargest(5, 'price')
top5
```

Задание 4


```python
max=authors_price.groupby('author_name')['price'].max()
min=authors_price.groupby('author_name')['price'].min()
mean=authors_price.groupby('author_name')['price'].mean()
authors_stat=pd.merge(min, max, on='author_name', how='inner')
authors_stat=pd.merge(authors_stat, mean, on='author_name', how='inner')
authors_stat=authors_stat.rename(columns={'price_x':'min_price', 'price_y':'max_price', 'price':'mean_price'})
authors_stat
```

Задание 5

```python
authors_price['cover']=['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info=pd.pivot_table(authors_price, values='price', index=['author_name'], columns=['cover'], aggfunc=np.sum, fill_value=0)
book_info
```
